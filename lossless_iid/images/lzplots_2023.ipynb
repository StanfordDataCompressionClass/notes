```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
window_sizes = [1, 10, 100, 1000, 10000, 100000, 1000000]
alice_sizes = [87555, 87376, 82599, 68778, 58605, 53074, 52965]

# Plotting the graph again
plt.figure(figsize=(10, 6))
plt.plot(window_sizes, alice_sizes, '-o', color='purple', markerfacecolor='purple', markersize=8)

# Labeling and formatting
plt.xlabel('Window Size')
plt.ylabel('Compressed Size (Bytes)')
plt.title('Impact of Window Size on LZ77 Compressed Size for Alice in Wonderland')
plt.xscale('log')
plt.xticks(window_sizes, window_sizes)
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()

plt.show()
```
